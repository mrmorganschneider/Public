```python
import PIL
from PIL import Image
#from PIL import ImageEnhance
from PIL import ImageDraw
from PIL import ImageFont

# read image and convert to RGB
image=Image.open("readonly/msi_recruitment.gif")
image=image.convert('RGB')

# build a list of 9 images which have different brightnesses
#enhancer=ImageEnhance.Brightness(image)
images=[]
#for i in range(1, 10):
    #images.append(enhancer.enhance(i/10))

#begin code MSchneider

for i in range(1, 4):
    if i == 1:
        for j in range(1, 4):
            r,g,b = image.split()
            if j == 1: r = r.point(lambda i: i * 0.1) 
            elif j == 2: r = r.point(lambda i: i * 0.5)
            else: r = r.point(lambda i: i * 0.9)
            images.append(Image.merge('RGB', (r, g, b)))
            
    elif i == 2:
        for j in range(1, 4):
            r,g,b = image.split()
            if j == 1: g = g.point(lambda i: i * 0.1) 
            elif j == 2: g = g.point(lambda i: i * 0.5)
            else: g = g.point(lambda i: i * 0.9)
            images.append(Image.merge('RGB', (r, g, b)))
            
    else:
        for j in range(1, 4):
            r,g,b = image.split()
            if j == 1: b = b.point(lambda i: i * 0.1) 
            elif j == 2: b = b.point(lambda i: i * 0.5)
            else: b = b.point(lambda i: i * 0.9)
            images.append(Image.merge('RGB', (r, g, b)))
                

# create a contact sheet from different brightnesses
first_image=images[0]
contact_sheet=PIL.Image.new(first_image.mode, (first_image.width*3,(first_image.height+100)*3))
x=0
y=0

for img in images:
    # Lets paste the current image into the contact sheet
    contact_sheet.paste(img, (x, y) )
    # Now we update our X position. If it is going to be the width of the image, then we set it to 0
    # and update Y as well to point to the next "line" of the contact sheet.
    if x+first_image.width == contact_sheet.width:
        x=0
        y=y+first_image.height+100
    else:
        x=x+first_image.width

# resize and display the contact sheet
contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2) ))

fnt = ImageFont.truetype(font='readonly/fanwood-webfont.ttf', size=30)

txt = ImageDraw.Draw(contact_sheet)
txt.text((0,230),"Channel 0 intensity 0.1", font=fnt, fill=(25,255,255))
txt.text((0,510),"Channel 1 intensity 0.1", font=fnt, fill=(255,25,255))
txt.text((0,790),"Channel 2 intensity 0.1", font=fnt, fill=(255,255,25))

txt.text((first_image.width*.5,230),"Channel 0 intensity 0.5", font=fnt, fill=(128,255,255))
txt.text((first_image.width*.5,510),"Channel 1 intensity 0.5", font=fnt, fill=(255,128,255))
txt.text((first_image.width*.5,790),"Channel 2 intensity 0.5", font=fnt, fill=(255,255,128))

txt.text((first_image.width,230),"Channel 0 intensity 0.9", font=fnt, fill=(230,255,255))
txt.text((first_image.width,510),"Channel 1 intensity 0.9", font=fnt, fill=(255,230,255))
txt.text((first_image.width,790),"Channel 2 intensity 0.9", font=fnt, fill=(255,255,230))

display(contact_sheet)
```
