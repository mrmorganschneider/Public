```python
import PIL
basewidth = 400
wpercent = (basewidth / float(image.size[0]))
hsize = int((float(image.size[1]) * float(wpercent)))
image = image.resize((basewidth, hsize), PIL.Image.ANTIALIAS)
image.save('resized_krusty.jpg')
display(image)
```

```python
text = pytesseract.image_to_string(Image.open('resized_krusty.jpg')) 
print(text)
```
